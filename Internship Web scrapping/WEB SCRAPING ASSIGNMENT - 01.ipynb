### 1) Write a python program to display all the header tags from wikipedia.org

```python
from bs4 import BeautifulSoup
import requests
```

```python
def header(url):
    page = requests.get(url)
    print(page)
    soup_wiki = BeautifulSoup(page.content)
    titles = []
    for i in soup_wiki.find_all(['h1','h2','h3','h4','h5','h6']):
        titles.append(i.text.replace('\n',''))
    print('List all the header tags :', *titles, sep='\n')
```

```python
header('https://en.wikipedia.org/wiki/Main_Page')
```

### 2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.


```python
def Movies(url1,url2):
    imbd_page1 = requests.get(url1)
    imbd_page2 = requests.get(url2)
    
    print(imbd_page1,imbd_page2)
    
    soup_imbd1 = BeautifulSoup(imbd_page1.content)
    soup_imbd2 = BeautifulSoup(imbd_page2.content)
    
    movies = []
    for i in soup_imbd1.find_all('h3','lister-item-header'):
        movies.append(i.text.replace('\n','').strip('()').split("("))
    
    for j in soup_imbd2.find_all('h3','lister-item-header'):
        movies.append(j.text.replace('\n','').strip("()").split("("))
    
    name=[]
    release_year = []
    for k in range(len(movies)):
        name.append(movies[k][0])
        release_year.append(movies[k][1])
    
    movie_name =[]
    for i in name:
        movie_name.append(i.split('.')[1])
    
    rat=[]
    for l in soup_imbd1.find_all('div',class_= 'ratings-bar'):
        rat.append(l.text.split()[0])
    for l in soup_imbd2.find_all('div',class_= 'ratings-bar'):
        rat.append(l.text.split()[0])
    
    import pandas as pd
    dict = {'Name':movie_name,'Rating': rat,'Year of Release': release_year }
    df = pd.DataFrame(dict)
    return(df)
Movies('https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc','https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc&start=51&ref_=adv_nxt') 
```

## 3) Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. name, rating, year of release) and make data frame.

```python
def Indian_Movies(url1):
    imbd_page1=requests.get(url1)
    soup_imbd1 = BeautifulSoup(imbd_page1.content)
       
    movies = []
    for i in soup_imbd1.find_all('h3',class_ ='lister-item-header'):
        movies.append(i.text.replace('\n','').strip('()').split("("))
            
    name=[]
    release_year = []
    for k in range(len(movies)):
        name.append(movies[k][0])
        release_year.append(movies[k][1])
    
    movie_name =[]
    for i in name:
        movie_name.append(i.split('.')[1])
    
    rat=[]
    for l in soup_imbd1.find_all('div',class_= 'ipl-rating-widget'):
        rat.append(l.text.split()[0])
    
    import pandas as pd
    dict = {'Name':movie_name,'Rating': rat,'Year of Release': release_year }
    df = pd.DataFrame(dict)
    return(df)
```

```python
Indian_Movies('https://www.imdb.com/list/ls056092300/?sort=user_rating,desc&st_dt=&mode=detail&page=1')
```

## 4) Write a python program to scrape product name, price and discounts from https://meesho.com/bagsladies/pl/p7vbp

```python
def meesho(url):
    meesho = requests.get(url)
    print('response:',meesho)
    soup_meesho = BeautifulSoup(meesho.content)
    ladies_bags = []
    for i in soup_meesho.find_all('p',class_ = 'Text__StyledText-sc-oo0kvp-0 bWSOET NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS'):
        ladies_bags.append(i.text.replace("'", "").lower())
        
    price = []
    for i in soup_meesho.find_all('h5',class_ = 'Text__StyledText-sc-oo0kvp-0 hiHdyy'):
        price.append(i.text.replace('₹',''))
        
    discount = []
    for i in soup_meesho.find_all('span',class_ = 'Text__StyledText-sc-oo0kvp-0 lnonyH'):
            discount.append(i.text.replace('%','').strip('').split(" ")[0])
    print('length of each columns:',len(ladies_bags),len(price),len(discount))
    import pandas as pd
    df = pd.DataFrame({'Product-name':ladies_bags,'Price':price,'Discount':discount})
    return(df)
meesho('https://meesho.com/bags%02ladies/pl/p7vbp')
```

## 5) Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:
### a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
def ODI_RANKING_MENS(url):
    ODI_teams = requests.get(url)
    ODI = BeautifulSoup(ODI_teams.content)
    team_name = []
    for i in ODI.find_all('span',class_ ='u-hide-phablet'):
        team_name.append(i.text.replace("\n"," "))
        team_name = team_name[:10]
    records = []
    for i in ODI.find_all('tr',class_ ='rankings-block__banner'):
        records.append(i.text.replace("\n"," ").split()[4:])
    for j in ODI.find_all('tr',class_ ='table-body'):
        records.append(j.text.replace("\n"," ").split()[-3:])
        records =records[:10]
    matches = []
    points =[]
    rating =[]
    for k in range(len(records)):
        matches.append(records[k][0])
        points.append(records[k][1])
        rating.append(records[k][2])
    import pandas as pd
    df = pd.DataFrame({'Team':team_name,'Matches':matches,'Points':points,'Rating':rating})
    return(df)

ODI_RANKING_MENS('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
```

### b) Top 10 ODI Batsmen along with the records of their team and rating.

```python
def ODI_top10_bats(url):
    Batsman = requests.get(url)
    print(Batsman)
    ODI_batsman = BeautifulSoup(Batsman.content)
    
    Batsman_name = []
    for i in ODI_batsman.find_all('div',class_ ='rankings-block__banner--name-large'):
        Batsman_name.append(i.text.replace("\n"," "))
    
    for j in ODI_batsman.find_all('td',class_ ='table-body__cell rankings-table__name name'):
        Batsman_name.append(j.text.replace("\n"," "))
        Batsman_name = Batsman_name[:10]
        
    team= []
    for i in ODI_batsman.find_all('div',class_ ='rankings-block__banner--nationality'):
        team.append(i.text.replace("\n",""))
    for j in ODI_batsman.find_all('span',class_ ='table-body__logo-text'):
        team.append(j.text.replace("\n"," "))
        team=team[:10]
        
    Rating=[]
    for i in ODI_batsman.find_all('div',class_ ='rankings-block__banner--rating'):
        Rating.append(i.text.replace("\n",""))
    for j in ODI_batsman.find_all('td',class_ ='table-body__cell rating'):
        Rating.append(j.text.replace("\n"," "))
        Rating=Rating[:10]
        
    dict = {'Batsman name':Batsman_name,'Team':team,'Rating':Rating}
    import pandas as pd
    df = pd.DataFrame(dict)
    return(df)

ODI_top10_bats('https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting')
```

# 7) Write a python program to scrape details of all the posts from coreyms.com. Scrape the heading, date, contentand the code for the video from the link for the youtube video from the post.

```python
def coreyms(url):
    web_tube = requests.get(url)
    print(web_tube)
    web_youtube = BeautifulSoup(web_tube.content)
    
    heading = []
    for i in web_youtube.find_all('h2',class_ = 'entry-title'):
        heading.append(i.text)
    heading.remove( 'Update (2019-09-03)')
    
    date = []
    for i in web_youtube.find_all('time',class_ = 'entry-time'):
        date.append(i.text)
    date.remove('September 3, 2019')
    
    content = []
    for i in web_youtube.find_all('div',class_ ='entry-content'):
        content.append(i.text.replace('\n',''))
    content.remove('Hey everyone. I wanted to give you an update on my videos. I will be releasing videos on threading and multiprocessing within the next week. Thanks so much for your patience. I currently have a temporary recording studio setup at my Airbnb that will allow me to record and edit the threading/multiprocessing videos. I am going to be moving into my new house in 10 days and once I have my recording studio setup then you can expect much faster video releases. I really appreciate how patient everyone has been while I go through this move, especially those of you who are contributing monthly through YouTube ')
    
    
    youtube_url=[]
    for i in web_youtube.find_all('iframe',class_ ="youtube-player"):
        youtube_url.append(i['src']) #for url we have to give source also
        
    dic = {'heading':heading,'date':date,'content':content,'url':youtube_url}
    import pandas as pd
    df = pd.DataFrame(dic)
    return(df)
coreyms('https://coreyms.com/')
```

### 9) Write a python program to scrape mentioned details from dineout.co.in :
#### i) Restaurant name
#### ii) Cuisine
#### iii) Location
#### iv) Ratings
#### v) Image URL

```python

def dineout(url):
    dineout = requests.get(url)
    print(dineout)
    dineout_soup = BeautifulSoup(dineout.content)
    
    restaurant_name = []
    for i in dineout_soup.find_all('div',class_ ='restnt-info cursor'):
        restaurant_name.append(i.text)
        
    cuisine = []
    for i in dineout_soup.find_all('span',class_ ='double-line-ellipsis'):
        cuisine.append(i.text.replace('₹','').split('|')[1])
    
    location=[]
    for i in dineout_soup.find_all('div',class_ ='restnt-loc ellipsis'):
        location.append(i.text)
        
    rating = []
    for i in dineout_soup.find_all('div',class_='restnt-rating rating-4'):
        rating.append(i.text)

    image_url = []
    for i in dineout_soup.find_all('img',class_='no-img'):
        image_url.append(i['data-src'])
    
    import pandas as pd
    dic = {'Restaurant name':restaurant_name,'Cuisine':cuisine,'Location':location,'Rating':rating,'image_url':image_url}
    df = pd.DataFrame(dic)
    return(df)
dineout('https://www.dineout.co.in/delhi-restaurants/buffet-special')
```

### 10) Write a python program to scrape first 10 product details which include product name , price , Image URL from
https://www.bewakoof.com/women-tshirts?ga_q=tshirts .

```python
'https://www.bewakoof.com/women-tshirts?ga_q=tshirts '
def bewakoof(url):
    tshirt = requests.get('https://www.bewakoof.com/women-tshirts?ga_q=tshirts ')
    print(tshirt)
    tshirt_soup = BeautifulSoup(tshirt.content)
    product_name = []
    for i in tshirt_soup.find_all('div',class_="productCardDetail"):
        product_name.append(i.text.split('₹')[0])

    price = []
    tribe_member = []
    for i in tshirt_soup.find_all('div',class_='productPriceBox clearfix'):
        price.append(i.text.split('₹')[1].replace('FEW LEFT',""))
        tribe_member.append(i.text.split('₹')[-1].replace('For TriBe Members',''))
    image_url = []
    for i in tshirt_soup.find_all('img',class_='productImgTag'):
        image_url.append(i['src'])
    image_url
    dic = {'Product Name':product_name,'Price':price,'Tribe member price':tribe_member,'image_url':image_url}
    import pandas as pd
    df = pd.DataFrame(dic)
    return(df)
bewakoof('https://www.bewakoof.com/women-tshirts?ga_q=tshirts ')
```
